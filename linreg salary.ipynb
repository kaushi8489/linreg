```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import seaborn as sns
```

# IMPORTING THE DATA

```python
sl = pd.read_csv('Salary_Data.csv')
print(sl)
```

# LINE PLOT

```python
sns.lmplot(x = 'YearsExperience',y = 'Salary',data = sl)
```

```python
x = sl[['YearsExperience']]
```

```python
y = sl['Salary']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(sl[['YearsExperience']] , sl['Salary'])
```

```python
lin_reg.predict([[5.1]])
```

```python
linreg.predict([[5.1]])
```

```python
y_pred = linreg.predict(x_test)
y_pred
```

```python
y_test
```

FINDING THE COEFFICIENTS AND INTERCEPTS

```python
# COEFFOCOENT
print(linreg.coef_)
```

```python
# INTERCEPT
print(linreg.intercept_)
```

EVALUATING THE MODEL

```python
# CALCULATING THE R SQUARED ERROR
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
print('Variance Score: %2f' %r2_score(y_test,y_pred))
```
